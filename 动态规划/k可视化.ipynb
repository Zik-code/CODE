```python
import matplotlib.pyplot as plt
import numpy as np

models = ['OmniAnomaly', 'MTAD_GAT', 'GDN', 'CAE_M', 'MAD_GAN', 'USAD', 'DAGMM', 'TranAD', 'DTAAD', 'ours']
auc_means = [0.957975, 0.9182, 0.955425, 0.959325, 0.9502, 0.9596, 0.9373, 0.9685, 0.970025, 0.976975]
f1_means = [0.9184, 0.870225, 0.89415, 0.889825, 0.914925, 0.914225, 0.89195, 0.94075, 0.933425, 0.957725]
p_means = [0.904275, 0.8774, 0.860725, 0.84205, 0.900775, 0.89575, 0.862375, 0.94005, 0.926975, 0.9446]
r_means = [0.948475, 0.8709, 0.950375, 0.960975, 0.948175, 0.948025, 0.9427, 0.9497, 0.950075, 0.9733]

x = np.arange(len(models))
width = 0.2

fig, ax = plt.subplots()
rects1 = ax.bar(x - width*1.5, auc_means, width, label='AUC')
rects2 = ax.bar(x - width/2, f1_means, width, label='F1')
rects3 = ax.bar(x + width/2, p_means, width, label='P')
rects4 = ax.bar(x + width*1.5, r_means, width, label='R')

ax.set_ylabel('')
ax.set_title('')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

fig.tight_layout()
plt.show()
```
